Write a function that takes a list as an argument, separates odd and even numbers into separate lists within it, and returns these lists.

```python
l = [2,13,18,93,22]
def func(list):
     even_list = []
     odd_list = []
     for i in list:
          if i % 2 == 0:
               even_list.append(i)
          else:
               odd_list.append(i)
     return even_list,odd_list

print(func(l))
```

The names of the students who achieved success in the engineering and medical faculties are given in the list below. The first three students represent the ranking of the engineering faculty, while the last three students belong to the medical faculty ranking. Print the student rankings by faculty using enumerate.

```python
students = ["Ali", "Veli", "Ayse", "Talat", "Zeynep", "Ece"]

for i,x in enumerate(students):
    if i<3:
        i += 1
        print("Mühendislik Fakültesi", i, ". öğrenci:",x)
    else:
        i -= 2
        print("Tıp Fakültesi", i, ".öğrenci:", x)
```

```python
students = ["Ali", "Veli", "Ayse", "Talat", "Zeynep", "Ece"]
faculties = ["Mühendislik", "Mühendislik", "Mühendislik", "Tıp", "Tıp", "Tıp"]
for index, faculty in enumerate(faculties):
    print(f"{faculty} fakültesi {index % 3 + 1}. öğrenci: {students[index]}")
```

Aşağıda 3 adet liste verilmiştir. Listelerde sırası ile bir dersin kodu, kredisi ve kontenjan bilgileri yer almaktadır. Zip kullanarak ders bilgilerini bastırınız.


```python
ders_kodu = ["CMP1005", "PSY1001", "HUK1005", "SEN2204"]
kredi = [3, 4, 2, 4]
kontenjan = [30, 75, 150, 25]

for index, (kredi, ders_kodu, kontenjan) in enumerate(zip(kredi, ders_kodu, kontenjan)):
    print(f"kredisi {kredi} olan {ders_kodu} kodlu dersin kontenjanı {kontenjan} kişidir.")
```

```python
set1 = set(["data", "python"])
set2 = set(["data", "function", "qcut", "lambda", "python", "miuul"])
def group_difference(set1, set2):
    if set1.issuperset(set2):
        print(set2.intersection(set1))
    else:
        print(set2.difference(set1))
        
group_difference(set1, set2)        
```
